## Section 1-1.以原點為中心擴散的圓形
***
### -  所需套件
### $\;\;\;\;$* matplotlib.pyplot - 可繪製圖形  
### $\;\;\;\;$* numpy - 運用 array 進行運算

### -  n 決定要畫幾個圓圈在半徑為 (10, 20) 之間

### -  plt.cm.jet 產生 n 個色票

### -  使用迴圈，改變圓的大小、顏色

### -  每一圈在畫時，會以順時針的方式每30度添加一個 marker，使用的是30度的旋轉矩陣，向量經由矩陣相乘改變藉以改變散落角度。

*** 

## 注意事項

### ax.axis 一行指令可設定圖的上下左右界線，無須再同時使用ax.set_ylim和ax.set_xlim

### marker除了可以直接挑圖案外也有相當多細項可調：
### $\;\;\;\;$ 1.  markersize 調大小
### $\;\;\;\;$ 2.  markerfacecolor 是否填滿
### $\;\;\;\;$ 3.  markeredgecolor 調 marker 的外框顏色
### $\;\;\;\;$ 4.  markeredgewidth 調 marker 的外框大小

***

### 二維旋轉矩陣定義為
$$ \begin{bmatrix}
-cos(\theta) & sin(\theta) \\
sin(\theta) & cos(\theta) 
\end{bmatrix},\; \theta為想旋轉的角度$$
<br>


```python
# packages
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D

# setting
fig, ax = plt.subplots(figsize = (6, 6))
n = 6
mymarker = ["s", "*", "d", "D", "p", "o"]
rad = np.linspace(10, 20, n)
colors = plt.cm.jet(np.linspace(0,1,n))

# draw and calculate
for i in range(n):
    ax.add_patch(plt.Circle((0, 0), rad[i],\
         fill = None, color = colors[i], lw = 3))
    v = [rad[i], 0]
    angle30 = np.pi/6
    rot30 = np.array([[np.cos(angle30), -np.sin(angle30)], [np.sin(angle30), np.cos(angle30)]])
    for j in range(12):
        plt.plot(v[0], v[1], marker = mymarker[i], markersize = 13, markerfacecolor="None", 
            markeredgecolor=colors[i], markeredgewidth=2)
        v = v@rot30

# adjust figure and show
ax.grid(True)
ax.axis([-21, 21, -21, 21])
plt.show()
```

## Section 1-2. 利用函數決定圓數量和半徑和 marker 的角度
***
### - 所需套件
* Line2D - 用來看到所有markers的種類

### - def 可寫個函式決定畫圓的數量、半徑大小、角度和 marker 的種類

### - 如果 marker 未設定，每圈 marker 的決定是從所有 markers 中隨機抽出的。

### - 如果 marker 設定數量比圓數量不同，會自己 extend 到圓的個數。

### - 最後 n = 500 呈現是漸進彩虹的色彩

***
## 注意事項

### np.random.choice(x, n) x 是要抽的內容 n是要抽的數量
<br>

```python
# function for drawing many circle
def draw_many_circle(n = 10, radius = [10, 30], angle = 30, marker = []):
    
    fig, ax = plt.subplots(figsize = (6, 6))
    rad = np.linspace(radius[0], radius[1], n, endpoint = True)
    colors = plt.cm.jet(np.linspace(0,1,n))
    theta = (angle/360) * (2 * np.pi)
    rot = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
    markerscount = int(360/angle)

    if len(marker) == 0:
        markerslist = np.random.choice(list(Line2D.markers.keys()), n)
    else:
        markerslist = np.tile(marker, int(np.ceil(n/len(marker))))[:n]
    
        
    for i in range(n):
        ax.add_patch(plt.Circle((0, 0), rad[i],\
            fill = None, color = colors[i], lw = 3))
        v = [rad[i], 0]
        for j in range(markerscount):
            plt.plot(v[0], v[1], marker = markerslist[i], markersize = 13, markerfacecolor="None", 
                markeredgecolor=colors[i], markeredgewidth=2)
            v = v @ rot
    
    ax.grid(True)
    ax.axis("equal")
    plt.show()

# exert functions for different parameters
draw_many_circle()
draw_many_circle(n = 10, radius = [0, 50], angle = 45)
draw_many_circle(n = 500, radius = [10, 50], angle = 30, marker = ["s", "o", 1, 2])

```

## Section 2-2. 利用函式決定個數
<br>

```python
# function for the arbitrary number of rectangle and lines
def drawmany(n):
    fig, ax = plt.subplots(figsize = (6, 6))
    ax.set_aspect("equal")
    ax.axis([-6, 6, -6, 6])
    angle = (2 * np.pi)/n
    rot = np.array([[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]])
    linebox = np.array([[0, 0], [3, 0], [3, -1], [5, -1], [5, 1], [3, 1], [3, 0]])
    for i in range(n):
        plt.plot(linebox[:, 0], linebox[:, 1])
        linebox = linebox@rot 

# draw many rectangles and lines
drawmany(4)
drawmany(20)
drawmany(50)
```

```python
# setting
n = 6
x = np.linspace(0, n, 1000)
y = np.cos(x * 2 * np.pi) + 1 


# watch the wave  
plt.plot(x, y)
plt.show()

# draw the flower
theta = np.linspace(0, 2 * np.pi, 1000)
plt.polar(theta, y, alpha = 0)
plt.plot(0, 0, marker = "o", ms = 12, color = "red")

# erase the frame and fill the color
axes = plt.gca()
axes.fill(theta,y,'orange')
axes.set_yticklabels([])
axes.set_xticklabels([])
axes.grid(False)
axes.spines['polar'].set_visible(False)

# show
plt.show()
```

```python
from matplotlib import cm
def flowers(n = [6], size = [5], colormap = "RdPu", width = 1, dotsize = 10):
    cmap = cm.get_cmap(colormap)
    index = np.linspace(0.5, 1, len(n))
    colors = cmap(index)
    # setting

    for i in range(len(n)):
        x = np.linspace(0, n[i], 1000)
        amplitude = size[i]
        y = amplitude * np.cos(x * 2 * np.pi) + amplitude

        # draw the flower
        theta = np.linspace(0, width * 2 * np.pi, 1000)
        plt.polar(theta, y, alpha = 0)
        axes = plt.gca() 
        axes.fill(theta,y,color = colors[i])
        axes.set_theta_offset(np.pi/6)

    # erase the frame and fill the color
    plt.plot(0, 0, marker = "o", ms = dotsize, color = "orange")
    axes.set_yticklabels([])
    axes.set_xticklabels([])
    axes.grid(False)
    axes.spines['polar'].set_visible(False)

    # show
    plt.show()
flowers([5], [0.3], width = 2, dotsize = 50)
flowers([15, 11, 8, 4], [1, 0.9, 0.5, 0.3], width = 1, colormap = "spring")
```

## Section 2-1. 認識卡方分配
***

### 卡方分配 (Chi-Square Distribution) 為 Gamma 分配的特例 $$\chi^2(df) = Gamma(\frac{df}{2}, 2) $$

***
### - 以下呈現的不同自由度下的pdf和cdf

### - 自由度設定從一到五

### - 用 for 迴圈繪製

### - 可以看到增加自由度會使 p.d.f 右偏的情形消失

### - c.d.f. 則是可觀察出自由度越大 x 增加 y 靠近 1 的速度趨緩
***
## 注意事項

### 圖中中文 title 須設定 fontproperties="SimSun"

### legend 參數設定

#### loc -"擺放方位"
#### bbox_to_anchor - "圖上座標幾成的位置介於(0, 1)"
#### frameon = False 不要框框

## 靈感來源：[Wiki_Chi-squared](https://en.wikipedia.org/wiki/Chi-squared_test)
<br>


```python
###### probability density function
# packages
from scipy.stats import chi2

# setting
x = np.linspace(0, 8, 1000)
colors = ["black", "blue", "green", "red", "purple"]
dfs = 5

# draw
for k in range(dfs):
    df = k + 1
    chi = chi2.pdf(x, df = df)
    plt.plot(x, chi, lw = 2, label = "k=" + str(df), color = colors[k])

# adjust figure and show
plt.title("機率密度函數", fontproperties="SimSun",fontsize=15)
plt.legend(loc = "center right", bbox_to_anchor = (0.9, 0.6), frameon=False)
plt.axis([0, 8, 0, 1])
plt.show()

###### cumulative density function
# setting
y = np.linspace(0, 1, 1000, endpoint = True)

# draw
for k in range(dfs):
    df = k + 1
    chi = chi2.ppf(y, df = df)
    plt.plot(chi,  y, lw = 2, label = "k=" + str(df), color = colors[k])

# adjust figure and show
plt.title("累積分布函數", fontproperties="SimSun",fontsize=15)
plt.legend(loc = "center right", bbox_to_anchor = (0.9, 0.3), frameon=False)
plt.axis([0, 8, 0, 1])
plt.show()
```

## Section 2-2. 卡方與常態的關係
***
### - Let $Z \sim Normal(0,1)$. Then, $Z^2$ is $\chi^2(1)$

### - np.random.normal or chisquare 從常態和分布中抽出 10000 個樣本

### - Z2 為平方後的值

### - 我們將他畫 histogram 但不顯示 (alpha = 0)，主要是擷取一些圖裡的資料

### - 為了取每個 Bar 中間 x 座標的資料必須將兩端 bar 的位置相加再除以2取 x 座標的位置

### - 最後可畫出兩個資料的分布狀態

### - 兩條線幾乎相當貼合，也印證此 theorem。
***
## 注意事項

### np.random.normal loc 指的是 mean; scale 指的是標準差

### density=True 可看到占比大小

### np.insert(arr, obj, value) arr-想要放的東西, obj-位置, value-5

### y 的上限是訂在最大值加0.3。
<br>

```python
# sample from chi-square distribution and normal distribution
z = np.random.normal(loc = 0, scale = 1, size = 10000)
z2 = z**2
chi = np.random.chisquare(df = 1, size = 10000)

# obtain the information from histogram
z2hist, z2n, _ = plt.hist(z2, bins = 300, alpha = 0, density = True)
chihist, chin,_ = plt.hist(chi, bins = 300, alpha = 0, density = True)

# get the middle of x from two sides of the bar
chin = (chin[1:] + chin[:-1])/2
z2n = (z2n[1:] + z2n[:-1])/2

# plot and show
plt.plot(np.insert(z2n, 0, 0), np.insert(z2hist, 0, 0), color = "green", alpha = 0.6, label = r"$Z^2$")
plt.plot(np.insert(chin, 0, 0), np.insert(chihist, 0, 0), color = "red", alpha = 0.3, label = r"$\chi^2$")
plt.axis([-0.2, 5, 0, np.max([z2hist, chihist])+0.3])
plt.legend()
plt.ylabel("Density")
plt.xlabel("x")
plt.show()
```

## Section 2-3.卡方表與自由度及輸出Excel
***
### - 所需套件
### $\;\;\;\;$ * scipy.stats 的 chi2 - 為了要得到 inverse of CDF

### - DataFrame的製作方法
### $\;\;\;\;$ * 先將所需的所有值經由迴圈放到一個 1 x k 的array中
### $\;\;\;\;$ * 再 reshape 成二維的形式
### $\;\;\;\;$ * 最後轉成 DataFrame 套件

### - 先新增一欄 degree of freedom

### - 將 degree of freedom 設為 index 再輸出

### - 輸出時使用 pd.to_excel 語法
***
## 注意事項

### pd.set_option 是要讓 dataframe 裡面的浮點數設定在小數點第三位。

### set_index(inplace = True)直接替換原本的 df

### to_excel 的細部微調：float_format 設定 datatype,  sheetname excel 左下方的 sheet 名稱
<br>

```python
# packages
import pandas as pd
from scipy.stats import chi2 

# make a df list and a zero array for appending caculated results
F = np.array([0.995, 0.99, 0.975, 0.95, 0.9, 0.1, 0.05, 0.025, 0.01, 0.005]) # cumalative to F
chiarray = np.array([0])

# loop for calculating p-value under different df
for i in range(29):
    x = chi2.ppf(1 - F, i + 1)  
    chiarray = np.append(chiarray, x)

# reshape the list and change it into dataframe
chiarray = chiarray[1:].reshape(29, 10)
chitable = pd.DataFrame(chiarray, columns = F)

# set index of the dataframe
chitable["degree of freedom"] = np.linspace(1, 29, 29, endpoint = True)
chitable.set_index("degree of freedom", inplace = True)

# output the result
chitable.to_excel("D:\statistics_python\chisquare_table.xlsx",
        float_format = '%.3f', sheet_name = "卡方分配-累積分布函數與自由度之機率")
pd.set_option('display.float_format', lambda x: '%.3f' % x)
chitable.head()
```
